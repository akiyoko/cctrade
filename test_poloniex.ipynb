```python
% matplotlib inline

import time
from datetime import datetime

import matplotlib.pyplot as plt
import pandas as pd
import poloniex

pd.options.display.max_rows = 20
plt.style.use('ggplot')
```

```python
polo = poloniex.Poloniex(timeout=10, coach=True)
[(k, v['name']) for (k, v) in polo.returnCurrencies().items()]
```

```python
polo = poloniex.Poloniex(timeout=10, coach=True)
period = polo.MINUTE * 5  # polo.DAY
tick = 100
chart_usdt_btc = polo.returnChartData('USDT_BTC', period=period, start=time.time() - period * tick)
chart_usdt_btc
```

```python
df = pd.DataFrame(chart_usdt_btc)
df['date'] = df['date'].apply(lambda x: datetime.fromtimestamp(x))
df['open'] = df['open'].astype(float)
df['close'] = df['close'].astype(float)
df['volume'] = df['volume'].astype(float)
# SMA (Simple Moving Average) is explained at https://coins.newbium.com/post/1873-indicator-on-poloniex-what-is-that
df['rmean'] = df['close'].rolling(window=50).mean()
df['rstd'] = df['close'].rolling(window=50).std()
df['upper_band'] = df['rmean'] + df['rstd'] * 2
df['lower_band'] = df['rmean'] - df['rstd'] * 2
# Delete columns
df = df.drop(['weightedAverage'], axis=1)
# df = df.sort_values(by=['date'], ascending=[True])
df.plot(x='date', y=['close', 'rmean', 'upper_band', 'lower_band'])
df
```

```python
df.describe()
```

```python
ax = df.plot(x='date', y='close')
df.plot(ax=ax, x='date', y='volume', secondary_y=True)
```

```python
df_latest = df.tail(5)
((df_latest['close'] - df_latest['rmean']) / df_latest['rstd']).values
```

```python
df_latest['volume'].values
```

```python
# std
'{:.2%}'.format(df_latest.tail(1)['rstd'].values[0] / df.tail(1)['close'].values[0])
```

```python
# LTC/BTC

currency_pair = 'BTC_{}'.format('LTC')
chart_ltc_btc = polo.returnChartData(currency_pair, period=polo.MINUTE * 5, start=time.time() - polo.MINUTE * 500, end=time.time())
df2 = pd.DataFrame(chart_ltc_btc)
df2['date'] = df2['date'].apply(lambda x: datetime.fromtimestamp(x))
df2['open'] = df2['open'].astype(float)
df2['close'] = df2['close'].astype(float)
df2['volume'] = df2['volume'].astype(float)
# SMA (Simple Moving Average) is explained at https://coins.newbium.com/post/1873-indicator-on-poloniex-what-is-that
df2['rmean'] = df2['close'].rolling(window=50).mean()
df2['rstd'] = df2['close'].rolling(window=50).std()
df2['upper_band'] = df2['rmean'] + df2['rstd'] * 2
df2['lower_band'] = df2['rmean'] - df2['rstd'] * 2
# Delete columns
df2 = df2.drop(['weightedAverage'], axis=1)
df2.plot(x='date', y=['close', 'rmean', 'upper_band', 'lower_band'])
df2
```

```python
df2_latest = df2.tail(5)
((df2_latest['close'] - df2_latest['rmean']) / df2_latest['rstd'].values).values
```

```python
df2_latest['volume'].values
```

```python
# std
'{:.2%}'.format(df2_latest.tail(1)['rstd'].values[0] / df2.tail(1)['close'].values[0])
```
